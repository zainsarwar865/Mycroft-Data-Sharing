# Generic Augmentations

```python
import matplotlib.pyplot as plt
import numpy as np

enola = np.asarray([81.3, 55, 55, 65])
random = np.asarray([20,40,40 ,50])
title = f"Generic Data Augmentation"
config = f"Generic_Data_Augmentation"

species = ("Dogs & Wolves","Food101-UPMC","Food101-ISIA 500","Imagenet-Tsinghua")

penguin_means = {
    'Enola': (enola),
    'Random sampling': (random),
}

#x = np.arange(len(species))  # the label locations
x = np.asarray([0, 0.1, 0.2, 0.3])
width = 0.03  # the width of the bars
multiplier = 0

fig, ax = plt.subplots()

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Accuracy on $D^{hard}$')
#ax.set_title('Generic Data Augmentation')
ax.set_xticks(x + width - 0.015, species, fontsize=7)
ax.legend(loc='upper right', ncols=3)
#ax.set_ylim(0, 1)
plot_name = config + ".jpg"
plot_path = os.path.join(plots_dir, plot_name)
plt.savefig(plot_path, dpi=300, bbox_inches='tight')

print(plot_path)
plt.show()
```

```python
imagenet_counts = []
adversarial_counts = []
benign_counts = []
ckpt = 1
ckpts = []
for path in l1[1:]:
    ckpts.append(ckpt)
    ckpt+=5
    df = pd.read_pickle(path)
    lens = len(df)
    #imagenet_cnt = len(df[df['data_type'] == 'imagenet_baseline']) / lens
    #benign_cnt = len(df[df['data_type'] == 'benign']) / lens
    adversarial_cnt = len(df[df['data_type'] == 'adversarial']) / lens
    #adversarial_cnt = len(df[df['data_type'] == 'adversarial']) / lens
    #imagenet_counts.append(imagenet_cnt)
    #benign_counts.append(benign_cnt)
    adversarial_counts.append(adversarial_cnt)
```

# UNICOM VS Features

```python
import matplotlib.pyplot as plt
import numpy as np

title = f"Unicom vs Feature Space Distances"
config = f"Unicom_featurespace_ablation"

enola = np.asarray([50, 13])
enola = enola / np.sum(enola)
random = np.asarray([4, 56])
random = random / np.sum(random)

species = ("Unicom", "Feature space distances")
penguin_means = {
    'Dogs & Wolves - Hard': (enola[0], random[0]),
    'Other': (enola[1], random[1]),
}

#x = np.arange(len(species))  # the label locations
x = np.asarray([0, 0.1])
width = 0.03  # the width of the bars
multiplier = 0

fig, ax = plt.subplots()

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    #ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('% in $D^{useful}$')
#ax.set_title('Unicom vs Feature Space Distances')
ax.set_xticks(x + width - 0.015, species)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 1)
#ax.set_title(title)
plot_name = config + ".jpg"
plot_path = os.path.join(plots_dir, plot_name)
plt.savefig(plot_path, dpi=300, bbox_inches='tight')

print(plot_path)
plt.show()
```
